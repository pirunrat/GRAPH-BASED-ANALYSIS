# Graph

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# Create a tiny 3x3 grayscale image for simplicity
image = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Create a graph
G = nx.Graph()

rows, cols = image.shape

# Add nodes with pixel intensity as an attribute
for i in range(rows):
    for j in range(cols):
        G.add_node((i, j), intensity=image[i, j])

        # Add edges to 4-neighbors (up, down, left, right)
        if i > 0:
            G.add_edge((i, j), (i-1, j))  # up
        if i < rows - 1:
            G.add_edge((i, j), (i+1, j))  # down
        if j > 0:
            G.add_edge((i, j), (i, j-1))  # left
        if j < cols - 1:
            G.add_edge((i, j), (i, j+1))  # right

# Print nodes with intensity
print("Nodes (with intensity):")
for node in G.nodes(data=True):
    print(node)

# Optional: visualize the graph
pos = {(i, j): (j, -i) for i in range(rows) for j in range(cols)}  # grid layout
nx.draw(G, pos, with_labels=True, node_color='lightgray', edge_color='black')
plt.title("Graph Representation of a 3x3 Image")
plt.show()
```
